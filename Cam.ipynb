```python
import cv2
import mtcnn
import numpy as np
from PIL import Image
def extractor(img):
    
    size=(100,100)
    detector=mtcnn.MTCNN()
    pixels=np.asarray(img)
    faces=detector.detect_faces(img)
    for res in faces:
        bounding_box = res['box']
        x=bounding_box[0]
        y=bounding_box[1]
        w=bounding_box[2]
        h=bounding_box[3]
        x1,y1=abs(x),abs(y)
        x2,y2=x1+w,y1+h
        face=pixels[y1:y2,x1:x2]
        image=Image.fromarray(face)
        image=image.resize(size)
        face_arra=np.asarray(image)
    return face_arra
    
def f(name,dir):
    cap=cv2.VideoCapture(0)
    
    count=0
    while True:
        ret,frame=cap.read()
        if extractor(frame) is not None:
            count=count+1
            face=extractor(frame)
            cv2.imwrite(dir+'/'+str(name)+'_'+str(count)+'.jpg',face)  # FOR COLOR IMAGES
            cv2.putText(face,str(count),(25,25),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
            cv2.imshow('Face Cropper',face)
        else:
            print("Face not Found")
            pass
        if cv2.waitKey(1)==5 or count==5:
            break
    cap.release()
    cv2.destroyAllWindows()
    print("Done")

name='Hari'
dir='images/'+name  
f(name,dir)
```

```python
import cv2
import numpy as np
import os
from tqdm import tqdm
import mtcnn

import pandas as pd
from PIL import Image
def Crete_Folder_Images(name,dir,tn):
    # Creating directory in the file
   
    detector=mtcnn.MTCNN()
    cam=cv2.VideoCapture(0)
    size =(100, 100)
    count=0
    for i in tqdm(range(1, 1+ tn)):
        ret,img=cam.read()
        pixels=np.asarray(img)
        faces=detector.detect_faces(img)
        count=count+1
        for res in faces:
            bounding_box = res['box']
            x=bounding_box[0]
            y=bounding_box[1]
            w=bounding_box[2]
            h=bounding_box[3]
            x1,y1=abs(x),abs(y)
            x2,y2=x1+w,y1+h
            face=pixels[y1:y2,x1:x2]
            image=Image.fromarray(face)
            image=image.resize(size)
            face_array=np.asarray(image)
            
            cv2.imwrite(dir+'/'+str(name)+'_'+str(i)+'.jpg',face_array)  # FOR COLOR IMAGES
            cv2.putText(face_array,str(count),(25,25),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
            cv2.imshow('Face Cropper',face_array)
            
            cv2.waitKey(200)
        cv2.imshow('DAtaset Creator',img)
        cv2.waitKey(1)
    cam.release()
    cv2.destroyAllWindows()
    print('TASK COMPLETED')
name='Malavika'
dir='images/'+name  
Crete_Folder_Images(name,dir,20)
```
